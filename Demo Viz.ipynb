```python
# Visualization 3: Bar chart
width = 0.35
ind = np.arange(len(companies))

fig, ax = plt.subplots(figsize=(10, 5))
rects1 = ax.bar(ind - width/2, sentiment_scores, width, label='Sentiment-based')
rects2 = ax.bar(ind + width/2, stock_prices, width, label='Traditional')

ax.set_title('Impact of Sentiment Analysis vs Traditional Methods')
ax.set_xticks(ind)
ax.set_xticklabels(companies)
ax.legend()

plt.show()
```

```python
# Sample data
data = {
    'Company': ['Apple', 'Microsoft', 'Amazon', 'Google', 'Tesla', 'Facebook', 'NVIDIA', 'Netflix', 'Adobe', 'Oracle'],
    'Sentiment Score': [0.8, 0.3, -0.5, 0.6, -0.7, 0.2, 0.5, -0.2, 0.7, 0.1],
    'Stock Price': [150, 200, 3500, 2700, 800, 350, 600, 550, 500, 80],
    'Portfolio Weight': [0.15, 0.2, 0.1, 0.15, 0.05, 0.1, 0.05, 0.1, 0.05, 0.05]
}

df = pd.DataFrame(data)
```

```python
# Visualization 1: Heatmap
plt.figure(figsize=(14, 2))
sns.heatmap([df['Sentiment Score']], cmap='coolwarm', square=True, xticklabels=df['Company'], cbar_kws={'label': 'Sentiment Score'})
plt.title('Sentiment Scores of Companies')
plt.show()
```

```python
# Visualization 2: Scatterplot with regression line
plt.figure(figsize=(10, 5))
sns.regplot(x=df['Sentiment Score'], y=df['Stock Price'], marker='o', ci=None)
plt.xlabel('Sentiment Score')
plt.ylabel('Stock Price')
plt.title('Correlation Between Sentiment Scores and Stock Prices')
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Stacked bar chart
fig, ax = plt.subplots(figsize=(14, 6))

bottom = np.zeros(len(df['Company']))
for index, row in df.iterrows():
    ax.bar(df['Company'], row['Portfolio Weight'] * row['Sentiment Score'], bottom=bottom, label=row['Company'])
    bottom += row['Portfolio Weight'] * row['Sentiment Score']

ax.set_title('Impact of Sentiment Analysis on Portfolio Weights')
ax.set_ylabel('Sentiment-adjusted Portfolio Weight')
ax.legend(title='Companies', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Company': ['Apple', 'Microsoft', 'Amazon', 'Google', 'Tesla', 'Facebook', 'NVIDIA', 'Netflix', 'Adobe', 'Oracle'],
    'Sentiment Score': [0.8, 0.3, -0.5, 0.6, -0.7, 0.2, 0.5, -0.2, 0.7, 0.1]
}

df = pd.DataFrame(data)
```

```python
# Create a graph with companies as nodes
G = nx.Graph()
G.add_nodes_from(df['Company'])

# Connect all nodes with each other
G.add_edges_from([(company_1, company_2) for i, company_1 in enumerate(df['Company']) for company_2 in df['Company'][i+1:]])

# Use sentiment scores as node attributes
nx.set_node_attributes(G, df.set_index('Company')['Sentiment Score'].to_dict(), 'sentiment_score')
```

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Set node sizes and colors based on degree centrality and sentiment score
node_sizes = [degree_centrality[node] * 500 for node in G.nodes()]
node_colors = [G.nodes[node]['sentiment_score'] for node in G.nodes()]

# Create the network graph visualization
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G, seed=42)
```

```python
# Use scatter plot instead of nx.draw
plt.scatter(*zip(*pos.values()), s=node_sizes, c=node_colors, cmap='coolwarm', alpha=0.5)

# Add labels
for node, (x, y) in pos.items():
    plt.text(x, y, node, fontsize=10, ha='center', va='center')

plt.colorbar(label='Sentiment Score')
plt.title('Network Graph of Companies - Size and Color-coded by Sentiment Score')
plt.savefig('network_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def update_graph(day):
    plt.clf()

    sentiment_scores = data['Company Sentiment Scores'][day]
    sentiment_dict = dict(zip(data['Company'], sentiment_scores))

    # Set node sizes and colors based on sentiment scores
    node_sizes = [degree_centrality[node] * 500 for node in G.nodes()]
    node_colors = [sentiment_dict[node] for node in G.nodes()]

    # Create the network graph visualization
    pos = nx.spring_layout(G, k=0.15, seed=42)
    plt.scatter(*zip(*pos.values()), s=node_sizes, c=node_colors, cmap='coolwarm', alpha=0.5)

    # Add labels
    for node, (x, y) in pos.items():
        plt.text(x, y, node, fontsize=10, ha='center', va='center')

    plt.colorbar(label='Sentiment Score')
    plt.title(f'Network Graph of Companies - Day {day + 1}\nSize and Color-coded by Sentiment Score')
```
